
###Estructura de Redes cristalinas

Una red cristalina es una disposición ordenada y repetitiva de átomos en el espacio. Esta regularidad es lo que da lugar a las propiedades físicas y químicas de los materiales cristalinos, como los metales, minerales o semiconductores.


## Vector de red
Un vector de red es cualquier vector que conecta un punto de la red con otro punto equivalente (es decir, con la misma disposición atómica). Se expresa como una combinación lineal entera de tres vectores base:

$$\vec{R} = n_1 \vec{a}_1 + n_2 \vec{a}_2 + n_3 \vec{a}_3
$$

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros
N = 20             # Número de puntos
a = 1.0            # Distancia entre puntos
t1 = np.array([a, 0])  # Vector de traslación en 1D (a lo largo del eje X)

# Inicializar matriz de coordenadas
R = np.zeros((N, 2))  # Cada punto tiene coordenadas (x, y)

# Construcción de la red unidimensional
for i in range(N):
    R[i, :] = i * t1  # Solo se traslada en X

# Visualización
plt.figure(figsize=(8, 2))
plt.plot(R[:, 0], R[:, 1], 'bo')        # Puntos en azul
plt.plot(R[:, 0], R[:, 1], 'k--', alpha=0.3)  # Línea guía
plt.title("Red Unidimensional de Puntos")
plt.xlabel("x")
plt.yticks([])  # Ocultamos el eje Y porque todos los puntos están en y = 0
plt.axis('equal')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros
N = 10              # Número de puntos en dirección X
M = 10              # Número de puntos en dirección Y
a = 1.0             # Distancia entre puntos

# Vectores primitivos de la red cuadrada
a1 = np.array([a, 0])
a2 = np.array([0, a])

# Inicializar arreglo de coordenadas
R = np.zeros((N * M, 2))

# Generar la red cuadrada
for i in range(N):
    for j in range(M):
        k = i * M + j
        R[k, :] = i * a1 + j * a2

# Visualización
plt.figure(figsize=(6, 6))
plt.plot(R[:, 0], R[:, 1], 'ro')             # Puntos en rojo
plt.title("Red Cuadrada 2D")
plt.xlabel("x")
plt.ylabel("y")
plt.axis('equal')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros
q = 1/2
m = q * np.sqrt(3)
N = 20  # Debe ser de la forma 2 + 3*n
lambda_val = 10

# Vectores primitivos
b1 = np.array([m, -q])
b2 = np.array([m, q])

# Generar red triangular
R = []
for j in range(1, N+1):
    for i in range(1, N+1):
        R.append(j*b1 + i*b2)
R = np.array(R)
k = len(R)

# Selección de puntos en patrón 2:3 y 3:3 (hexagonal)
G2 = R[1:k:3]
G3 = R[2:k:3]

# Unir en una sola matriz intercalada
d = len(G2)
G = np.zeros((2*d, 2))
G[0::2, :] = G2
G[1::2, :] = G3

# Construcción de la matriz M
long = 2 * d
M = np.zeros((long, long))
Rt = np.zeros((long**2, 2))  # Matriz de restas
for n in range(long):
    for h in range(long):
        l = long * n + h
        diff = G[n] - G[h]
        Rt[l] = diff
        if round(np.linalg.norm(diff)) == 1:
            M[n, h] = 1
        else:
            M[n, h] = 0

# Diagonalización del sistema
E, V = np.linalg.eig(M)
Vp = V[:, long-1]  # Último vector propio

# Gráfica de amplitudes en cada punto
plt.figure(figsize=(8, 6))
sc1 = plt.scatter(G[0::2, 0], G[0::2, 1], c=Vp[0::2], cmap='jet', s=60)
plt.colorbar(sc1, label='Amplitud (pares)')
sc2 = plt.scatter(G[1::2, 0], G[1::2, 1], c=Vp[1::2], cmap='jet', s=60)
#plt.colorbar(sc2, label='Amplitud (impares)')
plt.title('Amplitudes del campo en cada guía')
plt.xlabel('x')
plt.ylabel('y')
#plt.grid(True)
plt.axis('equal')
plt.show()
```
